```python
import pandas as pd
```

```python
a=[[1,2,3],[4,5,6],[7,8,9]]
```

```python
b=pd.DataFrame(a)
```

```python
b
```

Note: We can add columns and idex in the dataframe:

```python
b=pd.DataFrame(a,columns=['col1','col2','col3'], index=['EX1','EX2','EX3'])
```

```python
b
```

Note: with iloc you can filter specific part of dataframe with its rows and columns

```python
b.iloc[2,2]
```

```python
b.iloc[0:2,1]
```

```python
b.iloc[:,0:2]
```

Note: you can also split part of a specific dataframe with the name of columns and raws.

```python
b.loc['EX1','col1']
```

```python
b.loc['EX1']
```

```python
b.loc[:,'col2']
```

Note: in the pandas we also have different methods

```python
b.replace(9,100)
```

```python
b.replace({2:200, 3:400})
```
